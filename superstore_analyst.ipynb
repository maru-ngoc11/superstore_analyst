```python
#Super store Analysis
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

#1.Data loading
superstore = pd.read_csv("data/raw/superstore.csv", encoding = 'latin1')
```

```python
#2.EDA
print(f"Shape: {superstore.shape}")
print("-" * 50)

print("\nColumns:")
print(superstore.columns.tolist())
print("-" * 50)

print("\nData Types:")
print(superstore.dtypes)
print("-" * 50)

print(f"\nDuplicated Rows: {superstore.duplicated().sum()}")
print("-" * 50)

print("\nMissing Values:")
print(superstore.isnull().sum())
```

```python
#3.Standardize column names
superstore.columns = (superstore.columns.str.strip().str.lower().str.replace(" ", "_"))
```

```python
#4.Convert date columns
superstore['order_date'] = pd.to_datetime(superstore['order_date'], format='%m/%d/%Y')
superstore['ship_date'] = pd.to_datetime(superstore['ship_date'], format='%m/%d/%Y')
```
